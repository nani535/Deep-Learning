# Data Agumentation

### Data Agumentation is used to create new images from our old image.
### new images like zoom,height_shift_range,horizontal_flip,rotation_range etc these parameters are used to create new images. 

```python
from keras.preprocessing.image import ImageDataGenerator,array_to_img,img_to_array,load_img
```

```python
img=load_img(r"C:\Users\user\Desktop\dog.jpg")
```

```python
datagen=ImageDataGenerator(zoom_range=0.5,horizontal_flip=True,rotation_range=90,fill_mode='nearest',
                          height_shift_range=25)
```

```python
img
```

```python
img=img_to_array(img)
```

```python
img
#data is converted into array
```

```python
img=img.reshape((1,)+img.shape)
```

```python
img
```

```python
i=1
for data in datagen.flow(img,batch_size=2,
                         shuffle=True,
                         save_format='jpg',
                         save_prefix='pelli',
                         save_to_dir='images',
                         seed=100):
    i+=1
    if i >=20:
        break
#save_format:-new image is saved in that format i.e jpg
#save_prefix:-the image is saved with pelli___
#save_t0_dir:-directory to save the images
```
